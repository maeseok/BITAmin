### 2Î≤à LSTM

```python
import torch
import torch.nn as nn

# ÏÑ§Ï†ï
batch_size = 1
seq_len = 5
input_size = 10
hidden_size = 50  # Define hidden size

# LSTM Ï†ïÏùò
lstm = nn.LSTM(input_size=input_size, hidden_size=hidden_size, batch_first=True)

# ÏûÖÎ†• ÏãúÌÄÄÏä§
input_seq = torch.randn(batch_size, seq_len, input_size)

# Ï¥àÍ∏∞ Ïà®Í≤®ÏßÑ ÏÉÅÌÉúÏôÄ ÏÖÄ ÏÉÅÌÉú
h0 = torch.zeros(1, batch_size, hidden_size)  # num_layers = 1
c0 = torch.zeros(1, batch_size, hidden_size)  # num_layers = 1

# LSTM Ï∂úÎ†• Í≥ÑÏÇ∞
output, (hn, cn) = lstm(input_seq, (h0, c0))

print("LSTM output:", output)
print("Final hidden state:", hn)
print("Final cell state:", cn)
```

#### 2-2Î≤à) Ïó¨Îü¨ Ï∏µÏùò LSTMÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÏãúÌÄÄÏä§ Îç∞Ïù¥ÌÑ∞Î•º Ï≤òÎ¶¨ÌïòÎäî ÏΩîÎìúÎ•º ÏûëÏÑ±ÌïòÏÑ∏Ïöî. LSTMÏùò Ï∏µ ÏàòÏôÄ Ï¥àÍ∏∞ ÏÉÅÌÉúÎ•º ÏÑ§Ï†ïÌïòÎäî Î∂ÄÎ∂ÑÏùÑ Ï±ÑÏö∞ÏÑ∏Ïöî.(10Ï†ê)

```python
import torch
import torch.nn as nn

# ÏÑ§Ï†ï
batch_size = 1
seq_len = 7
input_size = 15
hidden_size = 64
num_layers = 3  # Number of layers in the LSTM

# Îã§Ï∏µ LSTM Ï†ïÏùò
lstm = nn.LSTM(input_size=input_size, hidden_size=hidden_size, num_layers=num_layers, batch_first=True)

# ÏûÖÎ†• ÏãúÌÄÄÏä§
input_seq = torch.randn(batch_size, seq_len, input_size)

# Ï¥àÍ∏∞ Ïà®Í≤®ÏßÑ ÏÉÅÌÉúÏôÄ ÏÖÄ ÏÉÅÌÉú
h0 = torch.zeros(num_layers, batch_size, hidden_size)
c0 = torch.zeros(num_layers, batch_size, hidden_size)

# LSTM Ï∂úÎ†• Í≥ÑÏÇ∞
output, (hn, cn) = lstm(input_seq, (h0, c0))

print("LSTM output:", output)
print("Final hidden states for each layer:", hn)
print("Final cell states for each layer:", cn)
```

#### 2-3Î≤à) LSTMÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÏãúÌÄÄÏä§ Îç∞Ïù¥ÌÑ∞Î•º Ï≤òÎ¶¨Ìïú ÌõÑ, ÎßàÏßÄÎßâ ÌÉÄÏûÑÏä§ÌÖùÏùò Ï∂úÎ†•ÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Îëê Í∞úÏùò ÌÅ¥ÎûòÏä§Î°ú Î∂ÑÎ•òÌïòÎäî Î™®Îç∏ÏùÑ ÏûëÏÑ±ÌïòÏÑ∏Ïöî. LSTMÏùò Ï†ïÏùò, Linear Î†àÏù¥Ïñ¥Î•º Ï†ïÏùòÌïòÍ≥†, ÎßàÏßÄÎßâ ÌÉÄÏûÑÏä§ÌÖùÏùò Ï∂úÎ†•ÏùÑ Ï∂îÏ∂úÌïòÎäî Î∂ÄÎ∂ÑÏùÑ Ï±ÑÏö∞ÏÑ∏Ïöî.(10Ï†ê)

```python
import torch
import torch.nn as nn

# ÏÑ§Ï†ï
batch_size = 1
seq_len = 10
input_size = 20
hidden_size = 50
num_classes = 2

# LSTM Í∏∞Î∞ò Î∂ÑÎ•ò Î™®Îç∏ Ï†ïÏùò
class LSTMClassifier(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(LSTMClassifier, self).__init__()
        # Define LSTM layer
        self.lstm = nn.LSTM(input_size=input_size, hidden_size=hidden_size, batch_first=True)
        # Define Linear layer
        self.fc = nn.Linear(hidden_size, num_classes)

    def forward(self, x):
        # Initialize hidden and cell states
        h0 = torch.zeros(1, x.size(0), hidden_size)  # (num_layers, batch_size, hidden_size)
        c0 = torch.zeros(1, x.size(0), hidden_size)  # (num_layers, batch_size, hidden_size)
        
        # Forward pass through LSTM
        out, _ = self.lstm(x, (h0, c0))
        
        # Extract the output of the last time step
        out = out[:, -1, :]  # Last time step
        
        # Forward pass through Linear layer
        out = self.fc(out)
        return out

# Î™®Îç∏ Ïù∏Ïä§ÌÑ¥Ïä§ ÏÉùÏÑ±
model = LSTMClassifier(input_size, hidden_size, num_classes)

# ÏûÖÎ†• ÏãúÌÄÄÏä§
input_seq = torch.randn(batch_size, seq_len, input_size)

# Î™®Îç∏ Ï∂úÎ†•
output = model(input_seq)
print("Output:", output)
```

### 1-2. Seq2SeqÏùò Encoder Íµ¨ÌòÑ ÏΩîÎìúÏûÖÎãàÎã§. seq2seqÏùò encoder ÏûëÎèô Í≥ºÏ†ïÏóê ÎåÄÌï¥ Í∞ÑÎã®ÌïòÍ≤å ÏÑ§Î™ÖÌï¥Ï£ºÏÑ∏Ïöî. (5Ï†ê)


```python
# Encoder

import torch
import torch.nn as nn

class Encoder(nn.Module):
    def __init__(self, input_dim, emb_dim, hid_dim, n_layers, dropout):
        super().__init__()

        self.hid_dim = hid_dim
        self.n_layers = n_layers

        # embedding: ÏûÖÎ†•Í∞íÏùÑ emd_dim Î≤°ÌÑ∞Î°ú Î≥ÄÍ≤Ω
        self.embedding = nn.Embedding(input_dim, emb_dim)

        # embeddingÏùÑ ÏûÖÎ†•Î∞õÏïÑ hid_dim ÌÅ¨Í∏∞Ïùò hidden state, cell Ï∂úÎ†•
        self.rnn = nn.LSTM(emb_dim, hid_dim, n_layers, dropout=dropout)

        self.dropout = nn.Dropout(dropout)

    def forward(self, src):
        # sre: [src_len, batch_size]

        embedded = self.dropout(self.embedding(src))

        # initial hidden stateÎäî zero tensor
        outputs, (hidden, cell) = self.rnn(embedded)

        # output: [src_len, batch_size, hid dim * n directions]
        # hidden: [n layers * n directions, batch_size, hid dim]
        # cell: [n layers * n directions, batch_size, hid dim]

        return hidden, cell
```

Îãµ) EncoderÎäî ÏûÖÎ†• ÏãúÌÄÄÏä§Î•º ÏûÑÎ≤†Îî© Î≤°ÌÑ∞Î°ú Î≥ÄÌôòÌïú ÌõÑ, LSTMÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÏãúÌÄÄÏä§Ïùò Í∞Å ÌÉÄÏûÑÏä§ÌÖùÏóêÏÑú Ïà®Í≤®ÏßÑ ÏÉÅÌÉúÏôÄ ÏÖÄ ÏÉÅÌÉúÎ•º Í≥ÑÏÇ∞Ìï©ÎãàÎã§. Ïù¥ Í≥ÑÏÇ∞Îêú Ïà®Í≤®ÏßÑ ÏÉÅÌÉúÏôÄ ÏÖÄ ÏÉÅÌÉúÎäî DecoderÏùò Ï¥àÍ∏∞ ÏÉÅÌÉúÎ°ú Ï†ÑÎã¨ÎêòÏñ¥ ÏãúÌÄÄÏä§Ïùò Ï∂úÎ†•ÏùÑ ÏÉùÏÑ±ÌïòÎäî Îç∞ ÏÇ¨Ïö©Îê©ÎãàÎã§. ÎìúÎ°≠ÏïÑÏõÉÏùÑ ÌÜµÌï¥ Î™®Îç∏Ïùò ÏùºÎ∞òÌôî ÏÑ±Îä•ÏùÑ Ìñ•ÏÉÅÏãúÌÇµÎãàÎã§.

### 1-3. Seq2Seq Íµ¨ÌòÑ ÏΩîÎìúÏùò ÎπàÏπ∏ÏùÑ Ï±ÑÏõåÏ£ºÏÑ∏Ïöî. (10Ï†ê)

```python
import torch
import torch.nn as nn

class Decoder(nn.Module):
    def __init__(self, output_dim, emb_dim, hid_dim, n_layers, dropout):
        super().__init__()

        self.output_dim = output_dim
        self.hid_dim = hid_dim
        self.n_layers = n_layers

        # Îã®Ïñ¥ Ïù∏Îç±Ïä§Î•º ÏûÑÎ≤†Îî© Î≤°ÌÑ∞Î°ú Î≥ÄÌôòÌïòÎäî Embedding Î†àÏù¥Ïñ¥
        self.embedding = nn.Embedding(output_dim, emb_dim)

        # EmbeddingÏùÑ ÏûÖÎ†•ÏúºÎ°ú Î∞õÏïÑÏÑú hid_dim ÌÅ¨Í∏∞Ïùò hidden state, cell ÏÉÅÌÉúÎ•º Ï∂úÎ†•ÌïòÎäî LSTM
        self.rnn = nn.LSTM(emb_dim, hid_dim, n_layers, dropout=dropout)

        # hidden stateÎ•º output_dim Ï∞®ÏõêÏúºÎ°ú Îß§ÌïëÌïòÎäî Linear Î†àÏù¥Ïñ¥
        self.fc_out = nn.Linear(hid_dim, output_dim)

        self.dropout = nn.Dropout(dropout)

    def forward(self, input, hidden, cell):
        # input: [batch_size] ÌòïÌÉú
        # hidden: [n_layers, batch_size, hid_dim] ÌòïÌÉú
        # cell: [n_layers, batch_size, hid_dim] ÌòïÌÉú

        # inputÏùÑ [1, batch_size]Î°ú Î≥ÄÌôò (batch_first=TrueÎ•º Í≥†Î†§Ìïú ÌòïÌÉú)
        input = input.unsqueeze(0) # input: [1, batch_size]

        # ÏûÖÎ†•ÏùÑ ÏûÑÎ≤†Îî© Î≤°ÌÑ∞Î°ú Î≥ÄÌôò
        embedded = self.dropout(self.embedding(input)) # [1, batch_size, emb_dim]

        # LSTMÏùÑ ÌÜµÌï¥ hidden stateÏôÄ cell stateÎ•º ÏóÖÎç∞Ïù¥Ìä∏
        output, (hidden, cell) = self.rnn(embedded, (hidden, cell))
        # output: [seq_len, batch_size, hid_dim]
        # hidden: [n_layers, batch_size, hid_dim]
        # cell: [n_layers, batch_size, hid_dim]

        # ÎßàÏßÄÎßâ time stepÏùò Ï∂úÎ†•ÏùÑ Linear Î†àÏù¥Ïñ¥Î•º ÌÜµÌï¥ ÏµúÏ¢Ö ÏòàÏ∏°ÏúºÎ°ú Î≥ÄÌôò
        prediction = self.fc_out(output.squeeze(0)) # [batch_size, output_dim]

        return prediction, hidden, cell
```

```python
class Seq2Seq(nn.Module):
    def __init__(self, encoder, decoder, device):
        super().__init__()

        self.encoder = encoder
        self.decoder = decoder
        self.device = device

        # encoderÏôÄ decoderÏùò hid_dimÏù¥ ÏùºÏπòÌïòÏßÄ ÏïäÎäî Í≤ΩÏö∞ ÏóêÎü¨ Î©îÏÑ∏ÏßÄ
        assert encoder.hid_dim == decoder.hid_dim, \
            'Hidden dimensions of encoder and decoder must be equal'
        # encoderÏôÄ decoderÏùò n_layersÏù¥ ÏùºÏπòÌïòÏßÄ ÏïäÎäî Í≤ΩÏö∞ ÏóêÎü¨ Î©îÏÑ∏ÏßÄ
        assert encoder.n_layers == decoder.n_layers, \
            'Encoder and decoder must have equal number of layers'

    def forward(self, src, trg, teacher_forcing_ratio=0.5):
        # src: [src_len, batch_size]
        # trg: [trg_len, batch_size]

        batch_size = trg.shape[1]
        trg_len = trg.shape[0]  # ÌÉÄÍ≤ü ÏãúÌÄÄÏä§ Í∏∏Ïù¥
        trg_vocab_size = self.decoder.output_dim  # ÌÉÄÍ≤ü Ïñ¥Ìúò ÌÅ¨Í∏∞

        # decoderÏùò outputÏùÑ Ï†ÄÏû•ÌïòÍ∏∞ ÏúÑÌïú tensor
        outputs = torch.zeros(trg_len, batch_size, trg_vocab_size).to(self.device)

        # Ï¥àÍ∏∞ hidden stateÏôÄ cell state ÏÑ§Ï†ï
        hidden, cell = self.encoder(src)

        # Ï≤´ Î≤àÏß∏ ÏûÖÎ†•Í∞íÏùÄ <sos> ÌÜ†ÌÅ∞
        input = trg[0]

        for t in range(1, trg_len):  # <eos> Ï†úÏô∏ÌïòÍ≥† trg_len-1 ÎßåÌÅº Î∞òÎ≥µ
            output, hidden, cell = self.decoder(input, hidden, cell)

            # prediction Ï†ÄÏû•
            outputs[t] = output

            # teacher forcingÏùÑ ÏÇ¨Ïö©Ìï†ÏßÄ ÎßêÏßÄ Í≤∞Ï†ï
            teacher_force = random.random() < teacher_forcing_ratio

            # Í∞ÄÏû• ÎÜíÏùÄ ÌôïÎ•†ÏùÑ Í∞ÄÏßÑ Îã®Ïñ¥ ÏñªÍ∏∞
            top1 = output.argmax(1)

            # teacher forcingÏùò Í≤ΩÏö∞Ïóê Îã§Ïùå ÏûÖÎ†•ÏúºÎ°ú target tokenÏùÑ ÏÇ¨Ïö©ÌïòÍ≥†, Í∑∏Î†áÏßÄ ÏïäÏúºÎ©¥ top1ÏùÑ ÏÇ¨Ïö©
            input = trg[t] if teacher_force else top1

        return outputs
```

### 2-1. Multi-Head Attention (5Ï†ê)
Scaled Dot-Product AttentionÏùÄ Transformer Î™®Îç∏ÏóêÏÑú Multi-Head AttentionÏùò Í∏∞Î≥∏ Ïó∞ÏÇ∞ÏúºÎ°ú, ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞Ïùò Ï§ëÏöîÌïú Î∂ÄÎ∂ÑÏùÑ ÌïôÏäµÌïòÍ≥† Í∞ïÌôîÌïòÎäî Îç∞ ÏÇ¨Ïö©Îê©ÎãàÎã§.

TransformerÏùò Attention value Í≥ÑÏÇ∞ÏãùÏùÑ Ï∞∏Í≥†ÌïòÏó¨ Scaled Dot-Product AttentionÏùò Íµ¨ÌòÑ ÏΩîÎìúÏùò ÎπàÏπ∏ÏùÑ Ï±ÑÏõåÏ£ºÏÑ∏Ïöî. (3Ï†ê)

```python
import torch
import torch.nn as nn
import math

class ScaleDotProductAttention(nn.Module):
    def __init__(self):
        super().__init__()
        self.softmax = nn.Softmax(dim=-1)

    def forward(self, q, k, v, mask=None):
        # scalingÏóê ÌïÑÏöîÌïú head_dim Í∞í ÏñªÍ∏∞
        # (batch_size, head, seq_len, head_dim)
        _, _, _, head_dim = q.size()

        # 1. KÎ•º transposeÌïòÍ∏∞ (seq_len, head_dimÏùò ÌñâÎ†¨ Ï†ÑÌôò)
        k_t = k.transpose(-1, -2)

        # 2. QÏôÄ K^TÏùò MatMul
        # (batch_size, head, q_seq_len, k_seq_len)
        attention_score = torch.matmul(q, k_t)

        # 3. Scaling
        attention_score /= math.sqrt(head_dim)

        # 4. MaskÍ∞Ä ÏûàÎã§Î©¥ ÎßàÏä§ÌÇπÎêú Î∂ÄÏúÑ -1e10ÏúºÎ°ú Ï±ÑÏö∞Í∏∞
        if mask is not None:
            attention_score = attention_score.masked_fill(mask == 0, -1e10)

        # 5. ÏÜåÌîÑÌä∏Îß•Ïä§ Ìï®Ïàò Ï∑®ÌïòÍ∏∞
        attention_score = self.softmax(attention_score)

        # 6. Attention Í≤∞Í≥ºÏôÄ VÏùò MatMul Í≥ÑÏÇ∞ÌïòÍ∏∞
        result = torch.matmul(attention_score, v)

        return result, attention_score
```

Îã§ÏùåÏùÄ Multihead AttentionÏùò Í≥ÑÏÇ∞ÏãùÏûÖÎãàÎã§.

Ïù¥Î•º Ï∞∏Í≥†ÌïòÏó¨ Multihead AttentionÏùò Íµ¨ÌòÑ ÏΩîÎìúÎ•º ÏûëÏÑ±Ìï¥Ï£ºÏÑ∏Ïöî. (2Ï†ê)

```python
import torch
import torch.nn as nn

class MultiHeadAttention(nn.Module):
    def __init__(self, d_model, head):
        super().__init__()
        self.d_model = d_model
        self.head = head
        self.head_dim = d_model // head  # Í∞Å Ìó§ÎìúÏùò Ï∞®Ïõê

        self.w_q = nn.Linear(d_model, d_model)
        self.w_k = nn.Linear(d_model, d_model)
        self.w_v = nn.Linear(d_model, d_model)
        self.w_o = nn.Linear(d_model, d_model)
        self.attention = ScaleDotProductAttention()

    def forward(self, q, k, v, mask=None):
        #  [batch_size, seq_len, d_model]
        batch_size, _, _ = q.size()

        # 1. Q, K, VÎ•º d_model Ï∞®ÏõêÏúºÎ°ú projection
        q, k, v = self.w_q(q), self.w_k(k), self.w_v(v)

        # 2. Q, K, VÎ•º head ÏàòÎßåÌÅº Î∂ÑÎ¶¨Ìï¥Ï£ºÍ∏∞
        # -> (batch_size, head, seq_len, head_dim)
        q = q.view(batch_size, -1, self.head, self.head_dim).transpose(1, 2)
        k = k.view(batch_size, -1, self.head, self.head_dim).transpose(1, 2)
        v = v.view(batch_size, -1, self.head, self.head_dim).transpose(1, 2)

        # 3. Scaled Dot-Product Attention ÏàòÌñâÌïòÍ∏∞
        out, attention_score = self.attention(q, k, v, mask)

        # 4. Î∂ÑÎ¶¨Îêú headÎì§ÏùÑ concat ÌïòÍ∏∞
        # -> (batch_size, seq_len, d_model)
        out = out.transpose(1, 2).contiguous().view(batch_size, -1, self.d_model)

        # 5. d_model Ï∞®ÏõêÏúºÎ°ú projection
        out = self.w_o(out)

        return out, attention_score
```

Îã§Ïùå ÏΩîÎìúÎäî PositionWiseFCFeedForwardNetwork Íµ¨ÌòÑ ÏΩîÎìúÏûÖÎãàÎã§.   
ÏΩîÎìúÎ•º Ïã§ÌñâÌï¥Ï£ºÏÑ∏Ïöî.

```python
class PositionWiseFCFeedForwardNetwork(nn.Module):
    def __init__(self,d_model,d_ff):
        super().__init__()
        self.w_1 = nn.Linear(d_model,d_ff)
        self.relu = nn.ReLU()
        self.w_2 = nn.Linear(d_ff,d_model)

    def forward(self, x):
        # Linear Layer1
        x = self.w_1(x)

        # ReLU
        x = self.relu(x)

        # Linear Layer2
        x = self.w_2(x)

        return x
```

### 2-2. Positional Encoding (5Ï†ê)

Îã§Ïùå Positional Encoding Í≥ÑÏÇ∞ÏãùÏùÑ Ï∞∏Í≥†ÌïòÏó¨ ÎπàÏπ∏ÏùÑ Ï±ÑÏõåÏ£ºÏÑ∏Ïöî.  
image.png

```python
class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_len, device):
        super().__init__()

        # Positional Encoding Ï¥àÍ∏∞Ìôî
        # 1. ÎπÑÏñ¥ÏûàÎäî tensor ÏÉùÏÑ±
        # (max_len, d_model)
        self.P_E = torch.zeros(max_len, d_model, device=device)
        # ÌïôÏäµÎêòÎäî Í∞íÏù¥ ÏïÑÎãòÏúºÎ°ú requires_gradÏùÑ FalseÎ°ú ÏÑ§Ï†ï
        self.P_E.requires_grad = False

        # 2. pos (0~max_len) ÏÉùÏÑ± (row Î∞©Ìñ• => unsqueeze(dim=1))
        pos = torch.arange(0, max_len, dtype=torch.float, device=device).unsqueeze(dim=1)

        # 3. _2i (0~2i) ÏÉùÏÑ± (col Î∞©Ìñ•)
        # 2iÎäî step = 2Î•º ÌôúÏö©ÌïòÏó¨ iÏùò 2Î∞∞ÏàòÎ•º ÎßåÎì¶
        _2i = torch.arange(0, d_model, step=2, dtype=torch.float, device=device)

        # 4. Ï†úÏïàÎêú positional encoding ÏÉùÏÑ±
        # ÏßùÏàò Ïù∏Îç±Ïä§Ïóê ÎåÄÌï¥ sin Ìï®Ïàò Ï†ÅÏö©
        # ÌôÄÏàò Ïù∏Îç±Ïä§Ïóê ÎåÄÌï¥ cos Ìï®Ïàò Ï†ÅÏö©
        self.P_E[:, 0::2] = torch.sin(pos / 10000 ** (_2i / d_model))
        self.P_E[:, 1::2] = torch.cos(pos / 10000 ** (_2i / d_model))

    def forward(self, x):
        # x seq Í∏∏Ïù¥Ïóê ÎßûÏ∂∞ PE return
        # (seq_len, d_model)
        batch_size, seq_len = x.size()
        PE_for_x = self.P_E[:seq_len, :]

        return PE_for_x
```

### 2-3. Layers: Encoder & Decoder (10Ï†ê)

Îã§ÏùåÏùÄ Encoder LayerÏôÄ Encoder Íµ¨ÌòÑÏûÖÎãàÎã§.

Encoder ÏïÑÌÇ§ÌÖçÏ≤òÎ•º Ï∞∏Í≥†ÌïòÏó¨ ÎπàÏπ∏ÏùÑ Ï±ÑÏõåÏ£ºÏÑ∏Ïöî.
(5Ï†ê)

```python
class EncoderLayer(nn.Module):
    def __init__(self, d_model, head, d_ff, dropout):
        super().__init__()
        self.attention = MultiHeadAttention(d_model, head)
        self.layerNorm1 = nn.LayerNorm(d_model)

        self.ffn = PositionWiseFCFeedForwardNetwork(d_model, d_ff)
        self.layerNorm2 = nn.LayerNorm(d_model)

        self.dropout = nn.Dropout(p=dropout)

    def forward(self, x, padding_mask):
        # residual connectionÏùÑ ÏúÑÌï¥ Ïû†Ïãú Îã¥ÏïÑÎëîÎã§.
        residual = x

        # 1. multi-head attention (self attention)
        x, attention_score = self.attention(q=x, k=x, v=x, mask=padding_mask)

        # 2. add & norm
        x = self.dropout(x) + residual
        x = self.layerNorm1(x)

        residual = x

        # 3. feed-forward network
        x = self.ffn(x)

        # 4. add & norm (2. add & norm Ï∞∏Í≥†)
        x = self.dropout(x) + residual
        x = self.layerNorm2(x)

        return x, attention_score
```

```python
class Encoder(nn.Module):
    def __init__(self, n_input_vocab, d_model, head, d_ff, max_len, padding_idx, dropout, n_layers, device):
        super().__init__()

        # Embedding
        self.input_emb = nn.Embedding(n_input_vocab, d_model, padding_idx=padding_idx)
        self.pos_encoding = PositionalEncoding(d_model, max_len, device)
        self.dropout = nn.Dropout(p=dropout)

        # nÍ∞úÏùò encoder layerÎ•º listÏóê Îã¥Í∏∞
        self.encoder_layers = nn.ModuleList([
            EncoderLayer(d_model=d_model,
                         head=head,
                         d_ff=d_ff,
                         dropout=dropout)
            for _ in range(n_layers)
        ])

    def forward(self, x, padding_mask):
        # 1. ÏûÖÎ†•Ïóê ÎåÄÌïú input embedding, positional encoding ÏÉùÏÑ±
        input_emb = self.input_emb(x)
        pos_encoding = self.pos_encoding(x)

        # 2. add & dropout
        x = self.dropout(input_emb + pos_encoding)

        # 3. nÎ≤à EncoderLayer Î∞òÎ≥µÌïòÍ∏∞
        for encoder_layer in self.encoder_layers:
            x, attention_score = encoder_layer(x, padding_mask)

        return x
```

Îã§ÏùåÏùÄ Decoder LayerÏôÄ Decoder Íµ¨ÌòÑÏûÖÎãàÎã§.  
image.png  
Decoder ÏïÑÌÇ§ÌÖçÏ≥êÎ•º Ï∞∏Í≥†ÌïòÏó¨ ÎπàÏπ∏ÏùÑ Ï±ÑÏõåÏ£ºÏÑ∏Ïöî. (5Ï†ê)

```python
class DecoderLayer(nn.Module):
    def __init__(self, d_model, head, d_ff, dropout):
        super().__init__()
        self.attention1 = MultiHeadAttention(d_model, head)
        self.layerNorm1 = nn.LayerNorm(d_model)

        self.attention2 = MultiHeadAttention(d_model, head)
        self.layerNorm2 = nn.LayerNorm(d_model)

        self.ffn = PositionWiseFCFeedForwardNetwork(d_model, d_ff)
        self.layerNorm3 = nn.LayerNorm(d_model)

        self.dropout = nn.Dropout(p=dropout)

    def forward(self, x, memory, look_ahead_mask, padding_mask):
        # 1. masked multi-head attention (self attention)
        residual = x
        x, _ = self.attention1(q=x, k=x, v=x, mask=look_ahead_mask)

        # 2. add & norm
        x = self.dropout(x) + residual
        x = self.layerNorm1(x)

        # 3. multi-head attention (encoder-decoder attention)
        residual = x
        x, _ = self.attention2(q=x, k=memory, v=memory, mask=padding_mask)

        # 4. add & norm
        x = self.dropout(x) + residual
        x = self.layerNorm2(x)

        # 5. feed-forward network
        residual = x
        x = self.ffn(x)

        # 6. add & norm
        x = self.dropout(x) + residual
        x = self.layerNorm3(x)

        return x
```

```python
class Decoder(nn.Module):
    def __init__(self, n_output_vocab, d_model, head, d_ff, max_len, padding_idx, dropout, n_layers, device):
        super().__init__()

        # Output embedding
        self.output_emb = nn.Embedding(n_output_vocab, d_model, padding_idx=padding_idx)
        self.pos_encoding = PositionalEncoding(d_model, max_len, device)
        self.dropout = nn.Dropout(p=dropout)

        # n Í∞úÏùò decoder layer
        self.decoder_layers = nn.ModuleList([DecoderLayer(d_model=d_model,
                                                         head=head,
                                                         d_ff=d_ff,
                                                         dropout=dropout)
                                             for _ in range(n_layers)])

    def forward(self, x, memory, look_ahead_mask, padding_mask):
        # 1. ÏûÖÎ†•Ïóê ÎåÄÌïú output embedding, positional encoding ÏÉùÏÑ±
        # (batch_size, seq_len, d_model)
        output_emb = self.output_emb(x)
        # (seq_len, d_model)
        pos_encoding = self.pos_encoding(x)

        # 2. add & dropout
        # (batch_size, seq_len, d_model)
        x = self.dropout(output_emb + pos_encoding)

        # 3. nÎ≤à DecoderLayer Î∞òÎ≥µÌïòÍ∏∞
        for decoder_layer in self.decoder_layers:
            x = decoder_layer(x, memory, look_ahead_mask, padding_mask)

        return x
```

### 2-4. Transformer (5Ï†ê)
ÏúÑÏóêÏÑú Íµ¨ÌòÑÌïú classÎì§ÏùÑ Ïù¥Ïö©Ìïú ÏµúÏ¢Ö transformer Î™®Îç∏ÏûÖÎãàÎã§.  
ÏïÑÎûò ÏΩîÎìúÎ•º Ï∞∏Í≥†ÌïòÏó¨ Ï†ÑÏ≤¥ ÏïÑÌÇ§ÌÖçÏ≤òÏùò ÎπàÏπ∏ÏùÑ Ï±ÑÏõåÏ£ºÏÑ∏Ïöî.

```python
class Transformer(nn.Module):
    def __init__(self, n_input_vocab, n_output_vocab, d_model, head, d_ff, max_len, padding_idx, dropout, n_layers, device):
        super().__init__()
        self.padding_idx = padding_idx
        self.device = device

        # Encoder
        self.encoder = Encoder(n_input_vocab=n_input_vocab,
                               d_model=d_model,
                               head=head,
                               d_ff=d_ff,
                               max_len=max_len,
                               padding_idx=padding_idx,
                               dropout=dropout,
                               n_layers=n_layers,
                               device=device)

        # Decoder
        self.decoder = Decoder(n_output_vocab=n_output_vocab,
                               d_model=d_model,
                               head=head,
                               d_ff=d_ff,
                               max_len=max_len,
                               padding_idx=padding_idx,
                               dropout=dropout,
                               n_layers=n_layers,
                               device=device)

        # linear layer
        # (batch_size, seq_len, n_output_vocab)
        self.linear = nn.Linear(d_model, n_output_vocab)

    def forward(self, src, tgt):
        # 1. ÏûÖÎ†•Ïóê Îî∞Î•∏ mask ÏÉùÏÑ±
        padding_mask = self.make_padding_mask(src, src)
        enc_dec_padding_mask = self.make_padding_mask(tgt, src)
        look_ahead_mask = self.make_padding_mask(tgt, tgt) * self.make_look_ahead_mask(tgt)

        # 2. encoder
        memory = self.encoder(src, padding_mask)

        # 3. decoder
        output = self.decoder(tgt, memory, look_ahead_mask, enc_dec_padding_mask)

        # 4. linear layer
        output = self.linear(output)

        return output

    def make_padding_mask(self, q, k):
        # q,kÏùò size = (batch_size, seq_len)
        _, q_seq_len = q.size()
        _, k_seq_len = k.size()

        q = q.ne(self.padding_idx)  # padding tokenÏùÑ 0, ÎÇòÎ®∏ÏßÄÎ•º 1Î°ú ÎßåÎì§Ïñ¥Ï§å
        q = q.unsqueeze(1).unsqueeze(3) # (batch_size, 1, q_seq_len, 1)
        q = q.repeat(1,1,1,k_seq_len)   # (batch_size, 1, q_seq_len, k_seq_len)

        k = k.ne(self.padding_idx)
        k = k.unsqueeze(1).unsqueeze(2) # (batch_size, 1, 1, k_seq_len)
        k = k.repeat(1,1,q_seq_len,1)   # (batch_size, 1, q_seq_len, k_seq_len)

        # and Ïó∞ÏÇ∞
        # (batch_size, 1, q_seq_len, k_seq_len)
        mask = q & k

        return mask

    def make_look_ahead_mask(self, tgt):
        _, seq_len = tgt.size()

        # torch.tril Ìï®ÏàòÎ•º ÏÇ¨Ïö©ÌïòÏó¨ ÌïúÏπ∏Ïî© Î∞ÄÎ†§ÎÇòÎ©∞ ÎßàÏä§ÌÇπÏùÑ Ìï¥Ï§å
        # (seq_len, seq_len)
        mask = torch.tril(torch.ones(seq_len,seq_len)).type(torch.BoolTensor).to(self.device)

        return mask
```
