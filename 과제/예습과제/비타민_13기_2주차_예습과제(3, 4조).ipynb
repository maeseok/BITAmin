### 2.3 (4점)

````python
import lightgbm as lgb
import logging

# LightGBM 로거 설정
logging.getLogger('lightgbm').setLevel(logging.ERROR)
# LightGBM 데이터 세트 생성
lgb_train = '''YOUR CODE'''(X_train, y_train)

# LightGBM 모델 설정 및 학습
params = {'objective': 'multiclass', 'num_class': 3, 'metric': 'multi_logloss', 'verbose': -1}
bst = lgb.train(params, lgb_train, num_boost_round=100)

# 모델 평가
y_pred = np.argmax(bst.predict(X_test, num_iteration=bst.best_iteration), axis=1)
accuracy = np.mean(y_pred == y_test)
print(f"Accuracy: {accuracy}")
````
