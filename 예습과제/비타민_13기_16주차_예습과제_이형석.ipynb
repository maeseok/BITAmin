#### 1-1. 아래 Convolution(합성곱)의 정의를 보고 합성곱의 의미를 적어주세요. (5점)

$(f*g)(t) = \int_{-\infty}^{\infty} f(\tau)g(t-\tau)\, d\tau$

답) f(τ) 함수를 g(t−τ) 함수에 대해 시간축을 따라 이동시키면서, 두 함수의 곱을 적분하여 하나의 함수로 결합하는 과정

#### 1-2. CNN의 주요 용어인 Filter, Channel, Stride, Feature map에 대해 각각 간단하게 한 줄로 설명해주세요. (5점)

답)
- Filter: 입력 이미지에서 특정 패턴이나 특징을 추출하기 위해 사용되는 가중치 행렬입니다.

- Stride: 필터를 적용할 때, 필터가 한 번에 이동하는 픽셀 수로, 출력 크기에 영향을 줍니다.
- Feature map: 필터가 입력 데이터에 적용된 결과로 얻어지는 출력 맵으로, 입력 이미지에서 추출된 특징을 나타냅니다.
