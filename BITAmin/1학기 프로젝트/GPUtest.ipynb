```python
import numpy as np
import matplotlib.pyplot as plt

# 함수 정의
def custom_cdf(x):
    if x < -1:
        return 0
    elif -1 <= x < 0:
        return 0.3
    elif 0 <= x < 1:
        return 0.8
    else:
        return 1

# 그래프를 그리기 위해 x 값 범위 생성
x_values = np.linspace(-2, 2, 400)  # -2부터 2까지 400개의 점 생성

# 각 x 값에 대한 CDF 값 계산
cdf_values = [custom_cdf(x) for x in x_values]

# 그래프 그리기
plt.plot(x_values, cdf_values, label='Custom CDF', color='blue')
plt.xlabel('x')
plt.ylabel('CDF')
plt.title('Custom CDF')
plt.grid(True)
plt.legend()
plt.show()
```
