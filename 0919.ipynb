```python
import numpy as np
np.random.seed(42) 
Score = np.random.randint(100, size=(10,4))
Score
```

```python
#1번 문제
import matplotlib.pyplot as plt

def plot_scatter(x_data, y_data, x_label, y_label, marker_type):
    plt.scatter(x_data, y_data, marker=marker_type)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()

plot_scatter(Score[:,1], Score[:,0], 'English', 'Korean', '^')
```

```python
#2번 문제
def sub_bar(ax,data, subject, bins=10):
    ax.hist(data, bins=bins)
    ax.set_xlabel(subject)
    ax.set_ylabel('Count')
    
def plot_histograms(data, subjects, bins=10):
    fig, axs = plt.subplots(2, 2, figsize=(10,10))
    axs = axs.ravel()
    for i in range(len(subjects)):
        sub_bar(axs[i], data[:, i], subjects[i], bins)
    plt.show()
    
subjects = ['Korea','English','Math','Science']
plot_histograms(Score, subjects)
```

```python
#3번 문제 ax.pie(sizes-a,b,c,d 개수 ,labels=?)
def plot_pie(sizes, labels):
    fig, ax = plt.subplots()
    ax.pie(sizes, labels=labels)
    plt.show()
Avg = Score.mean(axis=1)
print(np.sort(Avg))
arrlist = [1,1,8]
label = ['B', 'C', 'D']
plot_pie(arrlist, label)
```

```python
#4번 문제
def plot_boxplot(data, subjects):
    fig, ax = plt.subplots()
    ax.boxplot(data)
    ax.set_xticklabels(subjects)
    plt.show()

subjects = ['Kor', 'Eng', 'Math', 'Science']
Arrlist = [Score[:, i] for i in range(4)]
plot_boxplot(Arrlist, subjects)
```
