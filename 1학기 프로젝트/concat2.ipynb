```python
import pandas as pd

# 원본 CSV 파일 경로
file_path = "0530_931111row.csv"

# CSV 파일을 데이터프레임으로 읽기
data = pd.read_csv(file_path)

# 데이터를 반으로 나누기
half_length = len(data) // 2
data_part1 = data.iloc[:half_length, :]
data_part2 = data.iloc[half_length:, :]

# 나눈 데이터를 각각 CSV 파일로 저장
data_part1.to_csv("0530_part1.csv", index=False)
data_part2.to_csv("0530_part2.csv", index=False)

print("CSV 파일을 반으로 나누어서 저장하였습니다.")
```
